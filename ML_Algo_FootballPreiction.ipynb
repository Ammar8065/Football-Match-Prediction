```python
## Computing Rolling Averages [We can use how Liverpool performed in their previous 3 or 4 mathches and then the model can train on that specific data]
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(5, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
## Adding some new columns to calculate rolling averages
cols = ["gf" , "ga" , "sh" , "sot" , "dist" , "fk" , "pk" , "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
def make_predictions(data, predictors):
    # Split data into train and test sets
    test = data[data["date"] > '2024-01-01']
    train = data[data["date"] < '2024-01-01']

    # Ensure predictors exist in the train dataset
    missing_cols = [col for col in predictors if col not in train.columns]
    if missing_cols:
        raise ValueError(f"The following predictors are missing from the dataset: {missing_cols}")
    
    # Fit the model
    rf.fit(train[predictors], train["target"])

    # Predict on the test set
    preds = rf.predict(test[predictors])

    # Combine predictions and actual results
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)

    # Calculate precision
    precision = precision_score(test["target"], preds)

    return combined, precision
```
